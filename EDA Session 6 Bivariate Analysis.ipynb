```python
#Read the data.
file_loc="Visadataset.csv"
visa_df=pd.read_csv("Visadataset.csv")
visa_df
```

**Categorical vs Categorical**

```python
#here output column is case_status
#consider 2 catogorical columns: continent and case_status
visa_df['continent'].value_counts()
#visa_df['continent'] == 'Asia'
```

```python
#How to know how many visas are approved in Asia(out of 16861)
con1=visa_df['continent'] == 'Asia'
con2=visa_df['case_status'] == 'Certified'
visa_df[con1&con2]
```

```python
#generalized code for all continents - no of certified and denied visas
labels=visa_df['continent'].value_counts().keys()
for i in labels:
    con1=visa_df['continent'] == i
    con2=visa_df['case_status'] == 'Certified'
    con3=visa_df['case_status'] == 'Denied'
    cert_con=con1&con2
    deny_con=con1&con3
    print(f'{len(visa_df[cert_con])} got visas from {i}')
    print(f'{len(visa_df[deny_con])} didnot get visas from {i}')
```

```python
#certified visas  from all continet dataframe
certified_df = pd.DataFrame()
labels=visa_df['continent'].value_counts().keys()
for i in labels:
    con1=visa_df['continent'] == i
    con2=visa_df['case_status'] == 'Certified'
    cert_con=con1&con2
    #merge() cannot be used as initially dataframe is empty, common columns will not be present to merge. Hence concat()
    certified_df=pd.concat([certified_df,visa_df[cert_con]])
certified_df
```

```python
#Denied visas  from all continet dataframe
denied_df = pd.DataFrame()
labels=visa_df['continent'].value_counts().keys()
for i in labels:
    con1=visa_df['continent'] == i
    con3=visa_df['case_status'] == 'Denied'
    deny_con=con1&con3
    denied_df=pd.concat([denied_df,visa_df[deny_con]])
denied_df
```

```python
labels=visa_df['continent'].value_counts().keys()
cert=[]
deny=[]
for i in labels:
    con1=visa_df['continent'] == i
    con2=visa_df['case_status'] == 'Certified'
    con3=visa_df['case_status'] == 'Denied'
    cert_con=con1&con2
    deny_con=con1&con3
    cert.append(len(visa_df[cert_con]))
    deny.append(len(visa_df[deny_con]))
        
labels,cert,deny
```

```python
#Continent column is set as an index for the dataframe 
pd.DataFrame(zip(labels,cert,deny),columns=['Continent','Certified','Denied']).set_index("Continent")
```

**Numerical vs Numerical**

$Scatter Plots:$
- plt.scatter
- takes 2 args for x and y axes
- Both variables are numerical
- It provides r/nship b/w 2 variables like
   - +ve r/n
   - -ve r/n
   - No r/n

```python
#Eg1
x=[1,2,3,4,5]
y=[11,12,13,14,15]
plt.scatter(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,y)
#positevely related
```

```python
# Eg2 
# y=x**2
x=[i for i in range(0,5)]
y=[i**2 for i in x ]
plt.scatter(x,y)
plt.plot(x,y)
#positevely related
```

```python
#y=x**2
x=[i for i in range(-4,5)]
y=[i**2 for i in x ]
plt.scatter(x,y)
plt.plot(x,y)
```

```python
#consider two numerical  columns from the visa_df dataframe and plot them using scatterplot 
c1=visa_df['prevailing_wage']
c2=visa_df['no_of_employees']
plt.scatter(c1,c2)
plt.plot(c1,c2)
#the plotting is constant. these columns are independent. ie., no relation we can assume this graph as not related to each other
```

- we have to express the relation using mathematical terms ie., in numbers. for that we use Pearson Correlation coefficient(r)
- r varies from -1 to 1
- if r varies from -1 to 0 -> negatively co-realted
-  if r varies from 0 to 1 -> postively co-realted
- if r = 0 -> No relation
- r=Covariance(x,y)/(SD(x)*SD(y))


```python
visa_df.corr(numeric_only = True)
##all numeric columns are independent
```
