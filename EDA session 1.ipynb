```python
#creating DataFrame using LISTS and adding new column with values
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),columns=['Names','Age'],index=['A','B','C'])
df1['City']=['Hyd','Banglore','Chennai']
df1
```

```python
#interchanging column values. 
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),columns=['Names','Age'],index=['A','B','C'])
df1['City']=['Hyd','Chennai','Banglore']
df1
```
