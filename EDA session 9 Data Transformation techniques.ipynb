```python
# Step 1 import required packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Step 2 Read the Data
#Taking some random exponential data  from numpy package
exp_data=np.random.exponential(size=1000)
exp_data[:10]
```

```python
plt.hist(exp_data,bins=30,label='Exponential')
plt.legend()
```

```python
norm_data=np.random.normal(size=1000)
plt.hist(norm_data,bins=30,label="Normal")
plt.legend()
```

**Log Transformation - np.log()**

```python
log_data=np.log(exp_data)
log_data[:10]
```

```python
exp_data[:10],log_data[:10]
```

```python
plt.hist(log_data,bins=40,label='Log data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='Log Transformation')
plt.legend()
```

**Reciprocal transformation - np.reciprocal()**

```python
#Taking some random exponential data  from numpy package
exp_data1=np.random.exponential(size=1000)
#reciprocal=1/any_number
rec_data=np.reciprocal(exp_data1)
plt.figure(figsize=(15,5))
plt.subplot(1,2,1).hist(exp_data1,bins=40,label='Exponential Data')
plt.legend()
plt.subplot(1,2,2).hist(rec_data,bins=40,label='Reciprocal Transformation Data')
plt.legend()

```

**Square Root Transformation - np.sqrt()**

```python
#squareroot = any_number**0.5

sqrt_data=np.sqrt(exp_data1)
plt.figure(figsize=(15,5))
plt.subplot(1,2,1).hist(exp_data1,bins=40,label='Exponential Data')
plt.legend()
plt.subplot(1,2,2).hist(sqrt_data,bins=40,label='Square Root Transformation Data')
plt.legend()
```

```python
import seaborn as sns
sns.histplot(sqrt_data)
```

**Power Transformation - sklearn package**

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
pt_data=pt.fit([exp_data1]).transform([exp_data1])
plt.hist(pt_data[0],bins=40,label='Power Transformation')
plt.legend()
```
