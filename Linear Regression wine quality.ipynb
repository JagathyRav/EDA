**Finding MSE using python using formulae**

- steps:
    - As we have y_test and y_predictions, Using for loop, we substract each y_test value w.r.t y_predictions
    - Then Square the each substractions 
    - Then using a sum variable, add each squared value to finally get sum of all squared values.
    - Divide the sum by len(y_test) or len(y_predictions)


```python
s=0
for i in range(len(y_test)):  #len of y_test or y_predictions will be same. we can use any of the lengths
    v1=(y_test.values[i]-y_predictions[i])  #y_test is a column
    v1=v1**2
    s=s+v1
MSE=(s/len(y_test))
MSE
    
```

**Method 3: Variance Threshold method**

- Sometimes the column values doesnoot vary much. THEN WE CAN REMOVE THOSE COLUMN.
- for eg; in Loan prediction: loan term 360 column
- Every data has some variance. 
- Based on variance, we can decide the data is either constant or varying.
- if variance is 0, the data is not varying. By default python will drop such features/columns threshold =0.
- Variance threshold is Hyper parameter. We can choose threshold as any value say 0.1,0.2 0.3 etc..
- Based on internal team discussion, we will drop those feature which are having <= threshold value.
- most of the time, data has only one variance.
- it is vailable in sklearn.feature selection.

**p-value**
- H0: There is no r/n b/w x and y
- H1: There is r/n b/w x and y
- If p<0.05 - Reject the Null Hypothesis
- If p>0.05 - Not Reject the Null Hypothesis ie., does not Include the  observations

**Forward selection and backward selection**
-check article link in class file.
- https://courses.lumenlearning.com/introstats1/chapter/model-selection/#:~:text=The%20forward%20selection%20strategy%20is,measured%20by%20adjusted%20R2).

- **Forward selection method**
    - Start with epmty set of features and start adding one feature at a time.
    - Iterate through feature and select one feature that improves the model performance
    - Continue adding features one by one until we meet the criteria.
- **Backward selection method**
    - Start with full set of features
    - Iterate through each feature and  by removing the feature which is having less impact on the model.
    - Continue the process, till the performance is stable
- Both these methods are complex in nature. They take more time. hence many will avoid these methods in general.


**Vaeiance inflation factor (VIF)**
- VIF will give the analysis of multicollinearity.
- In VIF we find R-square value of each input feature w.r.t to other input feature.
- We will keep one feature as ip and other features are considered as op in Regression equation.
- This will keep on iterating through all features.
- Every feature behaves as both ip and op.
- As we know every regression line give Rsquare value from that value will fine VIF.
- we will drop the features having VIF>5 or VIF>!) based on internal team discussion.
