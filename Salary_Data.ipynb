```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')  #this step prevents warnings display
```

```python
# read the dataset 

df = pd.read_csv("Salary_Data.csv")
df
```

```python
df.shape
```

```python
cols=df.columns
cols
```

**Standardization of Numerical data (Z-score)**

```python
standard_df=pd.DataFrame()
for i in cols:
    col=df[i]
    mean=col.mean()
    std=col.std()
    standard_df[i]=(col-mean)/std
standard_df
```

```python
standard_df.abs()
```

```python
plt.subplot(1,2,1)
plt.boxplot(standard_df["YearsExperience"],vert=False)
plt.subplot(1,2,2)
plt.boxplot(standard_df["Salary"],vert=False)
    
    
    
    
```

```python
x=standard_df.drop('Salary',axis=1)
y=standard_df["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,  # Input data
                                                  y,  # output data
                                                  random_state=1234, # it select random samples
                                                  test_size=0.30)


```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_predictions=LR.predict(x_test)
y_predictions
```

```python
from sklearn.metrics import r2_score,mean_squared_error
R2=r2_score(y_test,y_predictions)
MSE=mean_squared_error(y_test,y_predictions)
#MSE**(1/2)
RMSE=np.sqrt(MSE)
#accuracy_score(y_test,y_predictions) # it is a regression tech
print("R-sqaure:",R2)
print("MSE:",MSE)
print("RMSE:",RMSE)
```
