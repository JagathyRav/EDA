$CategoricaltoNumericConversion$
- ML algorithms development are done using mathematics only. 
    - As we cannot use categorical columns to implement ML algo.s, Those columns must be converted to Numerical data mandatorily.
    - below methods are used for such conversion:
      - map() method for manual conversion.
      - One hot encoder: pd.get_dummies
      - Label encoder suing Scikit learn package.

```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the data.
file_loc="Visadataset.csv"
visa_df=pd.read_csv(file_loc)
visa_df
```

**Method 1 - map()**

- Consider a categorical column and observe how many unique labels are present  in that column
- Consider 'case_status'.
- unique labels are 'Certified' and Denied
- now create a dictionary by providing labels as keys and any numerical values to those each keys.
- the dictionary wil be like {'Certified' : 1,'Denied':0}
_ then map the dictionary to the the column using map(). and apply it to the same column. 

```python
#getting unique labels
unique_labels1=list(visa_df['case_status'].unique())
unique_labels1
```

```python
#creating a dictionary
d=dict()
for i in range(len(unique_labels1)):
    d[unique_labels1[i]]=i
d
```

```python
#mapping dictionary with the column.
visa_df['case_status']=visa_df['case_status'].map(d)
visa_df.head()
```

```python
#Consider continet column
labels=visa_df['continent'].unique()
d=dict()
for i in range(len(labels)):
    d[labels[i]]=i
visa_df['continent']=visa_df['continent'].map(d)
visa_df    
```

```python
#Read the data.
file_loc="Visadataset.csv"
visa_df=pd.read_csv(file_loc)
visa_df
```

**Method 2 - np.where()**
- used when column has only 2 unique labels.
- takes 3 args - condition, true value, false value

```python
#Method 2
#using np.where
 labels1=visa_df['case_status'].unique()
con=visa_df['case_status']==labels1[0]
visa_df['case_status']=np.where(con,0,1)
visa_df
```

**Method3  One hot encoder - pd.get_dummies()**

```python
#Read the data.
file_loc="Visadataset.csv"
visa_df=pd.read_csv(file_loc)
#Dropping ID column before applying one hot encoder. Mandatory step.
visa_df.drop('case_id',axis=1,inplace=True)
#applying one hot encoder.
pd.get_dummies(visa_df,dtype='int')
```

**Method 4 - LabelEncoder() from Sickit Learn package (sklearn package)**

```python
#Read the data.
file_loc="Visadataset.csv"
visa_df=pd.read_csv(file_loc)

#Read or import the LabelEncoder() method from the package sklearn
from sklearn.preprocessing import LabelEncoder
#save the method()
le=LabelEncoder()   #creating instance to the class
#Appply fit transform to a single cat column
visa_df['continent']=le.fit_transform(visa_df['continent'] )
visa_df
```

```python
 #Read the data.
file_loc="Visadataset.csv"
visa_df=pd.read_csv(file_loc)

#Read or import the LabelEncoder() method from the package sklearn
from sklearn.preprocessing import LabelEncoder
#save the method()
le=LabelEncoder()
#Apply fit transform to all cat columns
cat_cols=visa_df.select_dtypes(include='object').columns
for i in cat_cols:
    visa_df[i]=le.fit_transform(visa_df[i]) 
visa_df
le.inverse_transform(visa_df['case_status'][:5])
```
