```python
from sklearn.impute import KNNImputer
impute_knn = KNNImputer(n_neighbors=5)
num_cols=telecom_df.select_dtypes(exclude='object').columns
for i in num_cols:
    telecom_df[i]=impute_knn.fit_transform(telecom_df[[i]])
```

**8 Convert the negative value of the column to positive values**

**9 Plot Box Plot for the specified data**

```python
revenue=telecom_df['Total Revenue']
plt.boxplot(revenue,vert=False)
```

```python
#While handling Outliers, If label >UB or <LB,replace with median directly, 
#step1 finging Q1,Median, Q3
q1=round(np.percentile(revenue,25),2)
q2=round(np.percentile(revenue,50),2)
q3=round(np.percentile(revenue,75),2)

#step2   finding IQR
iqr=q3-q1

#step3 finding max and min limits
ub=q3+(1.5*iqr)
lb=q3-(1.5*iqr)
#step4 finding outliers and replacing outliers with UB or LB respectively

con1=revenue>ub
con2=revenue<lb
con=con1|con2
revenue=np.where(con,q2,revenue)

plt.boxplot(revenue,vert=False)
```
